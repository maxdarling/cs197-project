```python
import random
def dense_data(n):
    '''
    Makes dense data. Parameter 'n' is the last number, generates data [1, N].
    Everything written to 'dense.txt'
    '''
    results = []
    for i in range(1, n+1):
        results.append(i)
    return results

def sparse_data(n):
    '''
    Makes sparse data. Parameter 'n' is the end range. (Inclusive)
    For complete data set, n should be 2^64 - 1.
    '''
    results = []
    for i in range(1, n+1):
        number = random.randrange(1, 2**64)
        results.append(number)
    return results 
```

```python
dense = dense_data(100)
```

```python
performance = []

for num in dense:
    performance.append(dense.count(num))
```

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt

dense = dense_data(100)
y_pos = np.arange(len(dense))

performance = []

for num in dense:
    performance.append(dense.count(num))

plt.bar(y_pos, performance, align='center', alpha=0.5)
plt.ylabel('Usage')
plt.title('Dense Data Distribution')

plt.show()
plt.savefig("Dense_data.png")
```

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt

sparse = sparse_data(100)
y_pos = np.arange(len(sparse))

performance = []

for num in sparse:
    performance.append(sparse.count(num))

plt.bar(y_pos, performance, align='center', alpha=0.5)
plt.ylabel('Usage')
plt.title('Sparse Data Distribution')

plt.show()
plt.savefig("Sparse_data.png")
```
